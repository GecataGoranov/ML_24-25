#### 7. Reverse an array in **TWO** ways: using 'the Python way' and using a NumPy function

```python
arr = np.array([1, 2, 3])

python_arr = arr.copy()

for i in range(int(len(python_arr) / 2)):
    tmp = python_arr[i]
    python_arr[i] = python_arr[len(python_arr) - 1 - i]
    python_arr[len(python_arr) - 1 - i] = tmp

python_arr
```

```python
np_arr = arr.copy()
np_arr = np_arr[::-1]

np_arr
```
